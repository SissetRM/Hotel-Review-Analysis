# Hotel Review Analysis & Rating Prediction

This notebook explores hotel review data using both natural language processing (NLP) techniques and machine learning.

The primary objective is to predict the star rating (1–5) of a hotel review based solely on the review text. This is formulated as a supervised learning problem, where the model learns patterns in language that correspond to user ratings.

The workflow includes:
- Loading and exploring the dataset
- Converting text data into numerical features using TF-IDF
- Training a classification model (Logistic Regression)
- Evaluating model performance using standard metrics
- Analysing misclassifications to understand limitations

This project demonstrates a complete end-to-end pipeline, from raw text data to predictive modelling, with a focus on understanding both the strengths and limitations of machine learning applied to human language.

## Load Dataset

In this step, the dataset is loaded into a pandas DataFrame.

The dataset contains:
- `Review`: the text written by users
- `Rating`: the star rating (1–5)

This forms the basis of a supervised learning problem where we aim to predict the rating from the review text.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\alexs\OneDrive\Documents\2-Data-Analysis\Hotel-Review-Analysis\data\tripadvisor_hotel_reviews.csv")
df.head()
```

## Define Features and Target

The dataset is split into:

- **Features (X):** the review text  
- **Target (y):** the rating (1–5)

This defines the learning task:
> Predict the rating given the review text.

```python
x = df["Review"]   # text
y = df["Rating"]   # target (1–5)
```

## Train/Test Split

The data is split into training and testing sets using an 80/20 split.

- The **training set** is used to train the model  
- The **test set** is used to evaluate performance on unseen data  

This helps assess how well the model generalises.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

## Text Vectorisation (TF-IDF)

Machine learning models cannot process raw text, so the review text must be converted into numerical features.

TF-IDF (Term Frequency–Inverse Document Frequency) is used to:
- Assign importance to words based on how frequently they appear in a review
- Reduce the importance of very common words

Bigrams (two-word combinations) are included to capture context:
- Example: "not good", "very clean"

This improves the model's ability to understand phrases rather than individual words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
                            max_features=5000,
                            ngram_range=(1, 2)
                            )

x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```

## Model Training (Logistic Regression)

A Logistic Regression model is trained on the vectorised text data.

This model learns relationships between words (and phrases) and review ratings.

For example:
- Positive words → higher ratings  
- Negative words → lower ratings  

Logistic Regression is used because it performs well on text classification tasks and supports multiple classes.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train_vec, y_train)
```

## Generate Predictions

The trained model is used to predict ratings for the test dataset.

These predictions will be compared against the true ratings to evaluate model performance.

```python
y_pred = model.predict(x_test_vec)
```

## Model Evaluation

The model is evaluated using a classification report, which includes:

- **Precision:** How accurate the predictions are for each class  
- **Recall:** How many actual instances of each class are correctly identified  
- **F1-score:** A balance between precision and recall  

This provides a more detailed view than accuracy alone, especially for imbalanced datasets.

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

**Model Accuracy:** ~64%

This indicates the model performs significantly better than random guessing (20% for 5 classes), but still struggles with nuanced cases.

## Misclassification Analysis

An example of an incorrect prediction is displayed to better understand model limitations.

This helps identify patterns in errors, such as:
- Confusion between similar ratings (e.g. 3 vs 4)
- Over-reliance on strongly positive or negative words

Analysing these cases provides insight into how the model interprets language.

```python
for i in range(len(y_test)):
    if y_test.iloc[i] != y_pred[i]:
        print("Review:", x_test.iloc[i])
        print("Actual:", y_test.iloc[i])
        print("Predicted:", y_pred[i])
        print("---")
        break
```

## Key Observations

The model performs better on extreme ratings (1-star and 5-star), where language is more clearly positive or negative.

Mid-range ratings, particularly 3-star reviews, are more difficult to predict due to:
- Mixed sentiment
- Ambiguous language (e.g. "ok", "fine", "good but...")

This highlights a common challenge in natural language processing: interpreting nuanced human judgement.
